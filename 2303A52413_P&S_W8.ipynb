```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

calorie_intake = [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000]
exercise_duration = [60, 50, 30, 90, 45, 20, 75, 95, 15, 10]
bmi = [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(calorie_intake, bmi, color='blue')
plt.title('Daily Calorie Intake vs BMI')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('BMI')

plt.subplot(1, 3, 2)
plt.scatter(exercise_duration, bmi, color='green')
plt.title('Exercise Duration vs BMI')
plt.xlabel('Exercise Duration')
plt.ylabel('BMI')

plt.subplot(1, 3, 3)
plt.scatter(calorie_intake, exercise_duration, color='red')
plt.title('Daily Calorie Intake vs Exercise Duration')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Exercise Duration')

plt.tight_layout()
plt.show()

calorie_bmi_corr, _ = pearsonr(calorie_intake, bmi)
exercise_bmi_corr, _ = pearsonr(exercise_duration, bmi)
calorie_exercise_corr, _ = pearsonr(calorie_intake, exercise_duration)

print("Correlation Coefficient (Calorie Intake vs BMI):", calorie_bmi_corr)
print("Correlation Coefficient (Exercise Duration vs BMI):", exercise_bmi_corr)
print("Correlation Coefficient (Calorie Intake vs Exercise Duration):", calorie_exercise_corr)

if calorie_bmi_corr > 0:
    print("Consuming more calories is positively correlated with higher BMI.")
else:
    print("Consuming more calories is negatively correlated with BMI.")

if exercise_bmi_corr < 0:
    print("Exercise duration is negatively correlated with BMI, indicating that more exercise leads to lower BMI.")
else:
    print("Exercise duration is positively correlated with BMI.")
```
