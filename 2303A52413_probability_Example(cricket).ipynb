```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/cricket_dataset.csv")
data.head()
```

```python
def plot_histogram(df,column,bins=10):
  plt.figure(figsize=(10,5))
  sns.histplot(df[column],bins=bins,kde=True,color='red')
  plt.title(f'Histogram of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

```python
def plot_scatter(df,x_column,y_column):
  plt.figure(figsize=(10,5))
  sns.scatterplot(x=df[x_column],y=df[y_column],color='blue')
  plt.title(f'Scatter Plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```

```python
def plot_bar_chart(df,x_column,y_column):
  plt.figure(figsize=(10,5))
  sns.barplot(x=df[x_column],y=df[y_column],palette='viridis')
  plt.title(f'Bar Chart of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```

```python
file_path=('/content/cricket_dataset.csv')
data=pd.read_csv(file_path)
```

```python
plot_histogram(data,'Runs')
plot_histogram(data,'Wickets')
```

```python
plot_scatter(data,'Matches','Runs')
plot_scatter(data,'Economy Rate','Wickets')
```

```python
plot_bar_chart(data,'Player','50s')
plot_bar_chart(data,'Player','100s')
```
